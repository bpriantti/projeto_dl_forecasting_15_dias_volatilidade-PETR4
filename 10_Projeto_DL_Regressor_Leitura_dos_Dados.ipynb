#import libs:

```python
#import libs:
import pandas as pd
import numpy as np
import os

#libs para visualization dos dados:
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as md
import plotly.express as px
import plotly.graph_objects as Dash
import plotly.graph_objects as go

#---:
import warnings
warnings.filterwarnings("ignore")
```

```python
import pickle

with open("lstm_vol.txt", "rb") as fp:
  b = pickle.load(fp)
```

```python
b
```

### Step - 01:

```python
a = b[3][0]['step4_sinal']
c = b[3][0]['step4_pred']

fig = go.Figure()

fig.add_trace(go.Scatter(name="pred", y=a,mode='markers'))
fig.add_trace(go.Scatter(name="real", y=c,mode='markers'))
fig.update_layout(height=500, width=1000,
                  title=dict(text="Volatility Forecasting: Step-04"))
fig.show()
```

```python
fig = go.Figure()

fig = px.scatter(x=c, y=a, trendline="ols")

fig.update_layout(height=500, width=1000,
                  title=dict(text="Predict x Sample: Step-04"))
fig.show()
```

```python
step_1 = b[0][0]
step_2 = b[1][0]
step_3 = b[2][0]
step_4 = b[3][0]
```

```python
step_4.columns = ['sinal','pred']
```

```python
b[3][1]
```

```python
c = pd.concat([step_1,step_2,step_3,step_4]).reset_index()['sinal']
v = pd.concat([step_1,step_2,step_3,step_4]).reset_index()['pred']
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(name="pred", y=v,mode='markers'))
fig.add_trace(go.Scatter(name="real", y=c,mode='markers'))
fig.update_layout(height=500, width=1000,
                  title=dict(text="Volatility Forecasting: "))
fig.show()
```

```python
fig = go.Figure()

fig = px.scatter(x=c, y=v, trendline="ols")

fig.update_layout(height=500, width=1000,
                  title=dict(text="Predict x Sample: Step-04"))
fig.show()
```
